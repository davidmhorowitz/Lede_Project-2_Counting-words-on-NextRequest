```python
import pandas as pd
```

```python
oca = pd.read_csv('oca_filtered_df.csv')
ttx = pd.read_csv('ttx_filtered_df.csv')
pw = pd.read_csv('pw_filtered_df.csv')
ph = pd.read_csv('ph_filtered_df.csv')
controller = pd.read_csv('controller_filtered_df.csv')
dt = pd.read_csv('dt_filtered_df.csv')
ac = pd.read_csv('ac_filtered_df.csv')
ec = pd.read_csv('ec_filtered_df.csv')
all_depts = pd.read_csv('final_total_requests.csv')
```

```python
all_depts.head()
```

```python
merge2 = all_depts.merge(ph,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge3 = merge2.merge(pw,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge3.head()
```

```python
merge3 = merge3.rename(columns={'0_x' : 'Total', '0_y' : "Public Health", '0' : "Public Works"})
```

```python
merge3.head()
```

```python
merge4 = merge3.merge(controller,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge5 = merge4.merge(ttx,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge6 = merge5.merge(oca,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge6 = merge6.rename(columns={'0_x' : 'Controller', '0_y' : 'TTX', '0' : "OCA"})
```

```python
merge7 = merge6.merge(dt,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge8 = merge7.merge(ac,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge9 = merge8.merge(ec,
               left_on='Unnamed: 0',
               right_on='Unnamed: 0',
               how='left')
```

```python
merge9.head(5)
```

```python
NextRequest_Table_DF = merge9.rename(columns={'Unnamed: 0' : 'Word', '0_x' : 'Tech', '0_y' : 'Arts', '0' : 'Ethics'})
```

```python
# Let's only count words with at least 10 characters.
NextRequest_Table_df = NextRequest_Table_DF[NextRequest_Table_DF['Total'] >= 100]
```

```python
NextRequest_Table_df.head()
```

```python
NextRequest_Table_df.to_csv('NextRequest_Table_df.csv')
```
